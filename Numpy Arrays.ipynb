```python
# Step 1: Import NumPy
import numpy as np

# Step 2: Create a 1D array representing years of experience
years_exp = np.array([1, 3, 5, 7, 10])
print("--- Years of Experience ---")
print(years_exp, "\n")

# Step 3: Create a 2D array representing sample salaries (in thousands)
salaries = np.array([[50, 60, 70], 
                     [80, 90, 100]])
print("--- Salary Matrix ---")
print(salaries, "\n")

# Step 4: Create an array of zeros and ones for placeholder analysis
zeros_array = np.zeros((2, 2)) # 2x2 zeros
ones_array = np.ones((2, 3))   # 2x3 ones
print("--- Zeros Array ---")
print(zeros_array, "\n")
print("--- Ones Array ---")
print(ones_array, "\n")
```

```python
import numpy as np
years_exp = np.array([1, 3, 5, 7, 10])

# Step 1: Element-wise addition
exp_plus_5 = years_exp + 5 # Add 5 years to all experience values
print("Years + 5:", exp_plus_5)

# Step 2: Element-wise multiplication
exp_times_2 = years_exp * 2 # Multiply all values by 2
print("Years * 2:", exp_times_2)

# Step 3: Dot product (simulate salary projections)
sample1 = np.array([1, 2, 3])
sample2 = np.array([4, 5, 6])
dot_result = np.dot(sample1, sample2)
print("Dot Product:", dot_result)
```

```python
import numpy as np
years_exp = np.array([1, 3, 5, 7, 10])
salaries = np.array([[50, 60, 70], 
                     [80, 90, 100]])

print("--- Array Indexing & Slicing ---\n")

# Step 1: Access an individual element (Python is 0-indexed)
print("First year of experience:", years_exp[0])

# Step 2: Slice arrays (First two elements of the first row)
print("First two salaries:", salaries[0, :2]) 

# Step 3: Access all rows for a specific column (The ':' means 'all rows')
print("Second column salaries:", salaries[:, 1])

# Step 4: Negative indexing (-1 always grabs the very last element)
print("Last year of experience:", years_exp[-1])
```

```python
import numpy as np

print("--- Array Reshaping & Manipulation ---\n")

# Step 1: Reshape 1D array into 2x3 matrix for batch analysis
reshaped_exp = np.reshape(np.arange(1, 7), (2, 3))
print("Reshaped Experience Array:\n", reshaped_exp, "\n")

# Step 2: Flatten 2D arrays back into 1D
flattened_exp = reshaped_exp.flatten()
print("Flattened Array:", flattened_exp, "\n")

# Step 3: Transpose example (swaps rows and columns)
print("Transposed Array:\n", reshaped_exp.T)
```

```python
import numpy as np

salaries = np.array([[50, 60, 70], 
                     [80, 90, 100]])

print("--- Broadcasting Example ---\n")

# Step 1: Add a bonus array to salaries using broadcasting
bonus = np.array([5, 10, 15])

salaries_with_bonus = salaries + bonus 

print("Original Salaries:\n", salaries, "\n")
print("Bonus Array:", bonus, "\n")
print("Salaries after bonus:\n", salaries_with_bonus)
```

```python
import numpy as np

years_exp = np.array([1, 3, 5, 7, 10])
salaries = np.array([[50, 60, 70], 
                     [80, 90, 100]])

print("--- Statistical Analysis ---\n")

# Step 1: Mean (average) of years of experience
print("Mean experience:", np.mean(years_exp))

# Step 2: Standard deviation of experience (how spread out the numbers are)
print("Std deviation of experience:", np.std(years_exp))

# Step 3: Max and Min salaries
print("Max salary:", np.max(salaries))
print("Min salary:", np.min(salaries))

# Step 4: Sum of salaries
print("Sum of all salaries:", np.sum(salaries))
```

```python
# Step 1: Apply trigonometric functions
angles = np.array([0, np.pi/4, np.pi/2])
print("Sine of angles:", np.sin(angles))
print("Cosine of angles:", np.cos(angles))
# Step 2: Apply function along rows (sum salaries per person)
salary_sums = np.apply_along_axis(np.sum, 1, salaries)
print("Sum of Salaries per person:", salary_sums)
```

```python
import pandas as pd
import numpy as np

print("--- Testing Groupby Logic on Dummy Data ---\n")

# Step 1: Load the dummy data you already have
df_data = pd.read_csv('modified_data.csv')
print("Loaded Dataset:\n", df_data.head(), "\n")

# Step 2: Select relevant columns (the ones that actually exist!)
df_subset = df_data[['X', 'Y', 'Sum_XY']]
print("Subset of Data:\n", df_subset.head(), "\n")

# Step 3: Clean the data
df_clean = df_subset.dropna().copy()

# Step 4: Categorize X into groups 
# (Assigns 'High' if X is 30 or greater, otherwise 'Low')
df_clean['X_Category'] = np.where(df_clean['X'] >= 30, 'High', 'Low')

# Step 5: Group data by the new category, compute average of Y
grouped_data = df_clean.groupby(['X_Category'])['Y'].mean()
print("Grouped Average of Y:\n", grouped_data.head(), "\n")

# Step 6: Reset index for readability
grouped_data = grouped_data.reset_index()
print("Formatted Grouped Data:\n", grouped_data.head())
```
