```python
# Step 1: Import NumPy
import numpy as np

# Step 2: Create a 1D array representing years of experience
years_exp = np.array([1, 3, 5, 7, 10, 12, 15])
print("--- Years of Experience ---")
print(years_exp, "\n")

# Step 3: Create a 2D array representing sample salaries (in thousands)
salaries = np.array([[50, 60, 70, 75, 80], 
                     [80, 90, 100, 110, 120]])
print("--- Salary Matrix ---")
print(salaries, "\n")

# Step 4: Create an array of zeros and ones for placeholder analysis
zeros_array = np.zeros((2, 2)) # 2x2 zeros
ones_array = np.ones((2, 3))   # 2x3 ones
print("--- Zeros Array ---")
print(zeros_array, "\n")
print("--- Ones Array ---")
print(ones_array, "\n")

# Step 5: Create a 3x3 identity matrix (Try It Challenge)
identity_matrix = np.eye(3) 
print("--- 3x3 Identity Matrix ---")
print(identity_matrix, "\n")
```

```python
import numpy as np
years_exp = np.array([1, 3, 5, 7, 10])

print("--- Original Array ---")
print("Years of Experience:", years_exp, "\n")

# Step 1: Element-wise addition
exp_plus_5 = years_exp + 5 
print("Years + 5:", exp_plus_5)

# Step 2: Element-wise multiplication
exp_times_2 = years_exp * 2 
print("Years * 2:", exp_times_2)

# Step 3: Dot product
sample1 = np.array([1, 2, 3])
sample2 = np.array([4, 5, 6])
dot_result = np.dot(sample1, sample2)
print("Dot Product:", dot_result, "\n")

# Step 4: Subtraction and Division
exp_minus_1 = years_exp - 1 # Subtract 1 from all values
print("Years - 1 (Subtraction):", exp_minus_1)

exp_div_2 = years_exp / 2 # Divide all values by 2
print("Years / 2 (Division):", exp_div_2, "\n")

# Step 5: Exponential and Logarithm
# np.exp() calculates e^x (Euler's number raised to the power of the array value)
exp_array = np.exp(years_exp) 
print("Exponential (np.exp):", exp_array)

# np.log() calculates the natural logarithm (ln) for each value
log_array = np.log(years_exp) 
print("Logarithm (np.log):", log_array)
```

```python
import numpy as np
years_exp = np.array([1, 3, 5, 7, 10])
salaries = np.array([[50, 60, 70], 
                     [80, 90, 100]])

print("--- Array Indexing & Slicing ---\n")

# Step 1: Access an individual element
print("First year of experience:", years_exp[0])

# Step 2: Slice arrays 
print("First two salaries:", salaries[0, :2]) 

# Step 3: Access all rows for a specific column
print("Second column salaries:", salaries[:, 1])

# Step 4: Negative indexing 
print("Last year of experience:", years_exp[-1], "\n")

# Step 5: Reverse arrays using slicing
reversed_years = years_exp[::-1]
print("Reversed Years of Experience:", reversed_years)

# Step 6: Slice 2D arrays for subgroups of salaries
salary_subgroup = salaries[:, 1:] 
print("\nSalary Subgroup (Last two columns):\n", salary_subgroup)

high_earners = salaries[1, :]
print("\nHigh Earners Subgroup (Second row only):", high_earners)
```

```python
import numpy as np

print("--- Array Reshaping & Manipulation ---\n")
base_array = np.arange(1, 7)

# Step 1: Reshape 1D array into 2x3 matrix
reshaped_exp = np.reshape(base_array, (2, 3))
print("Reshaped Experience Array (2x3):\n", reshaped_exp, "\n")

# Step 2: Flatten 2D arrays back into 1D
flattened_exp = reshaped_exp.flatten()
print("Flattened Array:", flattened_exp, "\n")

# Step 3: Transpose example (swaps rows and columns)
print("Transposed Array (3x2):\n", reshaped_exp.T, "\n")

# Step 4: Reshape into different dimensions for modeling
model_data_3x2 = np.reshape(base_array, (3, 2))
print("Reshaped for Modeling (3x2):\n", model_data_3x2, "\n")
model_data_6x1 = np.reshape(base_array, (6, 1))
print("Reshaped as a Column Vector (6x1):\n", model_data_6x1)
```

```python
import numpy as np

salaries = np.array([[50, 60, 70], 
                     [80, 90, 100]])

print("--- Broadcasting Multiplication ---\n")

print("Original Salaries:\n", salaries, "\n")

# --- Method 1: Scalar Broadcasting ---
scaling_factor = 1.1 
scaled_salaries_scalar = salaries * scaling_factor

print(f"1. Scaled by {scaling_factor} (Scalar Broadcasting):\n", scaled_salaries_scalar, "\n")


# --- Method 2: Array Broadcasting ---
tier_multipliers = np.array([1.0, 1.5, 2.0])
scaled_salaries_array = salaries * tier_multipliers

print("Tier Multipliers Array:", tier_multipliers, "\n")
print("2. Scaled by Array (Column Broadcasting):\n", scaled_salaries_array)
```

```python
import numpy as np

years_exp = np.array([1, 3, 5, 7, 10])
salaries = np.array([[50, 60, 70], 
                     [80, 90, 100]])

print("--- Advanced Statistical Analysis ---\n")

# Step 1: Compute the median (the middle value when sorted)
print("Median experience:", np.median(years_exp))
print("Median salary:", np.median(salaries), "\n")

# Step 2: Compute percentiles
print("50th Percentile of experience:", np.percentile(years_exp, 50))
print("75th Percentile of salaries:", np.percentile(salaries, 75), "\n")
print("25th Percentile of salaries (per column):")
print(np.percentile(salaries, 25, axis=0))
```

```python
import numpy as np
years_exp = np.array([1, 3, 5, 7, 10])
salaries = np.array([[50, 60, 70], 
                     [80, 90, 100]])

print("--- Array Transformations ---\n")

# Step 1: Use np.sqrt() and np.log()
sqrt_exp = np.sqrt(years_exp)
print("Square Root of Experience:\n", sqrt_exp, "\n")
log_salaries = np.log(salaries)
print("Natural Log of Salaries:\n", log_salaries, "\n")


print("--- Custom Functions with apply_along_axis ---\n")
def calculate_range(data_array):
    """Calculates the difference between the highest and lowest values in an array."""
    return np.max(data_array) - np.min(data_array)

# Step 3: Apply the custom function
salary_ranges_per_row = np.apply_along_axis(calculate_range, 1, salaries)
salary_ranges_per_col = np.apply_along_axis(calculate_range, 0, salaries)

print("Original Salaries Matrix:\n", salaries, "\n")
print("Salary Spread per Row (axis=1):", salary_ranges_per_row)
print("Salary Spread per Column (axis=0):", salary_ranges_per_col)
```

```python
import pandas as pd
import numpy as np

print("--- Testing Groupby Logic on Dummy Data ---\n")

# Step 1: Load the dummy data you already have
df_data = pd.read_csv('modified_data.csv')
print("Loaded Dataset:\n", df_data.head(), "\n")

# Step 2: Select relevant columns (the ones that actually exist!)
df_subset = df_data[['X', 'Y', 'Sum_XY']]
print("Subset of Data:\n", df_subset.head(), "\n")

# Step 3: Clean the data
df_clean = df_subset.dropna().copy()

# Step 4: Categorize X into groups 
# (Assigns 'High' if X is 30 or greater, otherwise 'Low')
df_clean['X_Category'] = np.where(df_clean['X'] >= 30, 'High', 'Low')

# Step 5: Group data by the new category, compute average of Y
grouped_data = df_clean.groupby(['X_Category'])['Y'].mean()
print("Grouped Average of Y:\n", grouped_data.head(), "\n")

# Step 6: Reset index for readability
grouped_data = grouped_data.reset_index()
print("Formatted Grouped Data:\n", grouped_data.head())
```
