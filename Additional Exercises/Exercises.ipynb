```python
import numpy as np
import pandas as pd

STUDENT_NAME = "Calvin"
STUDENT_ID = 250698

df = pd.read_csv('Most Popular Programming Languages.csv')
df.columns = df.columns.str.strip()

date_col = None
for col in df.columns:
    if col.lower() in ['date', 'month']:
        date_col = col
        break

if date_col is None:
    raise ValueError("No Date or Month column found. Check your CSV headers.")

df[date_col] = pd.to_datetime(df[date_col])
df.rename(columns={date_col: 'Date'}, inplace=True)

all_languages = sorted([col for col in df.columns if col != 'Date'])

last_three_digits = STUDENT_ID % 1000
language_index = last_three_digits % len(all_languages)
assigned_language = all_languages[language_index]

analysis_df = df[['Date', assigned_language]].copy()
analysis_df.rename(columns={assigned_language: 'Popularity', 'Date': 'Month'}, inplace=True)
analysis_df['Growth_Rate'] = analysis_df['Popularity'].pct_change() * 100
analysis_df['Moving_Avg'] = analysis_df['Popularity'].rolling(window=12).mean()
analysis_df['Moving_STD'] = analysis_df['Popularity'].rolling(window=12).std()

conditions = [
    (analysis_df['Growth_Rate'] > 5),
    (analysis_df['Growth_Rate'] < -5)
]
choices = ['Growth', 'Decline']

analysis_df['Phase'] = np.select(conditions, choices, default='Stable')
analysis_df.fillna(0, inplace=True)

stat_summary = analysis_df['Popularity'].describe().to_frame()
phase_counts = analysis_df['Phase'].value_counts()

initial_pop = analysis_df['Popularity'].iloc[0]
final_pop = analysis_df['Popularity'].iloc[-1]
overall_growth = ((final_pop - initial_pop) / initial_pop) * 100

print(f"Student Name: {STUDENT_NAME}")
print(f"Assigned Language: {assigned_language} (Index: {language_index})")
print("=" * 50)
print("\nDataFrame Output")
print(analysis_df[['Month', 'Popularity', 'Growth_Rate', 'Moving_Avg', 'Moving_STD', 'Phase']].head(10))
print("\nStatistical Summary")
print(stat_summary)
print("\nGrowth Phase Counts")
print(phase_counts)
print("\nKey Performance Metrics")
print(f"Initial Popularity: {initial_pop:.2f}%")
print(f"Final Popularity: {final_pop:.2f}%")
print(f"Overall Growth: {overall_growth:.2f}%")
print(f"Growth Periods: {phase_counts.get('Growth', 0)}")
print(f"Decline Periods: {phase_counts.get('Decline', 0)}")
print(f"Stable Periods: {phase_counts.get('Stable', 0)}")
```

```python
import numpy as np
import pandas as pd

STUDENT_NAME = "Calvin"
STUDENT_ID = 250698

# Step 2: Load dataset
df = pd.read_csv('Most Popular Programming Languages.csv')

# Clean column names
df.columns = df.columns.str.strip()

# Detect Month column
month_col = None
for col in df.columns:
    if col.lower() == 'month':
        month_col = col
        break

if month_col is None:
    raise ValueError("Month column not found")

df[month_col] = pd.to_datetime(df[month_col])
df.rename(columns={month_col: 'Month'}, inplace=True)

# Step 3: Assign languages
language_cols = sorted([col for col in df.columns if col != 'Month'])

last_three = STUDENT_ID % 1000
idx1 = last_three % len(language_cols)
idx2 = (idx1 + 1) % len(language_cols)

lang1 = language_cols[idx1]
lang2 = language_cols[idx2]

# Step 4: Prepare data
analysis_df = df[['Month', lang1, lang2]].copy()
analysis_df.rename(columns={lang1: 'Lang_A', lang2: 'Lang_B'}, inplace=True)

# Step 5: Growth rates
analysis_df['Growth_A'] = analysis_df['Lang_A'].pct_change() * 100
analysis_df['Growth_B'] = analysis_df['Lang_B'].pct_change() * 100

# Step 6: Rolling stats
analysis_df['MA_A'] = analysis_df['Lang_A'].rolling(window=6).mean()
analysis_df['MA_B'] = analysis_df['Lang_B'].rolling(window=6).mean()

analysis_df['STD_A'] = analysis_df['Lang_A'].rolling(window=6).std()
analysis_df['STD_B'] = analysis_df['Lang_B'].rolling(window=6).std()

analysis_df.fillna(0, inplace=True)

# Step 7: Central tendencies
mean_A = analysis_df['Lang_A'].mean()
mean_B = analysis_df['Lang_B'].mean()

median_A = analysis_df['Lang_A'].median()
median_B = analysis_df['Lang_B'].median()

std_A = analysis_df['Lang_A'].std()
std_B = analysis_df['Lang_B'].std()

# Step 8: Normalize using NumPy
norm_A = (analysis_df['Lang_A'] - np.mean(analysis_df['Lang_A'])) / np.std(analysis_df['Lang_A'])
norm_B = (analysis_df['Lang_B'] - np.mean(analysis_df['Lang_B'])) / np.std(analysis_df['Lang_B'])

# Step 9: Correlation
correlation = np.corrcoef(norm_A, norm_B)[0, 1]

# Step 10: Dominance ratio
count_A_dominant = np.sum(analysis_df['Lang_A'] > analysis_df['Lang_B'])
total_months = len(analysis_df)
dominance_ratio = (count_A_dominant / total_months) * 100

# Step 11: Relative Performance Index
rpi = np.mean(norm_A - norm_B)

# Step 12: Cross-over points
cross_over = analysis_df[(analysis_df['Lang_A'] > analysis_df['Lang_B']) &
                         (analysis_df['Lang_A'].shift() <= analysis_df['Lang_B'].shift())]

# Step 13: Summary DataFrame
summary_df = pd.DataFrame({
    'Mean_A': [mean_A],
    'Mean_B': [mean_B],
    'Median_A': [median_A],
    'Median_B': [median_B],
    'Std_A': [std_A],
    'Std_B': [std_B],
    'Correlation': [correlation],
    'Dominance_Ratio (%)': [dominance_ratio],
    'RPI': [rpi]
})

# Step 14: Outputs
print(f"Student Name: {STUDENT_NAME}")
print(f"Language A: {lang1}")
print(f"Language B: {lang2}")
print("=" * 50)

print("\nComparison Table (first 10 rows)")
print(analysis_df.head(10))

print("\nSummary Statistics")
print(summary_df)

print("\nDominance Info")
print(f"Months A > B: {count_A_dominant} out of {total_months}")
print(f"Dominance Ratio: {dominance_ratio:.2f}%")

print("\nCross-over Points (first few)")
print(cross_over[['Month', 'Lang_A', 'Lang_B']].head(10))

# Step 15: Interpretation
if dominance_ratio > 50:
    dominant_lang = lang1
else:
    dominant_lang = lang2

if std_A < std_B:
    stability = f"{lang1} is more consistent"
else:
    stability = f"{lang2} is more consistent"

print("\nInterpretation")
print(f"{dominant_lang} shows stronger dominance over time.")
print(stability)
print(f"Correlation between trends: {correlation:.3f}")

# Clean column names
df.columns = df.columns.str.strip()

# Detect Month column
month_col = None
for col in df.columns:
    if col.lower() == 'month':
        month_col = col
        break

if month_col is None:
    raise ValueError("Month column not found")

# Convert to datetime
df[month_col] = pd.to_datetime(df[month_col])
df.rename(columns={month_col: 'Month'}, inplace=True)

# Step 3: Assign language
language_cols = sorted([col for col in df.columns if col != 'Month'])

last_three = STUDENT_ID % 1000
index = last_three % len(language_cols)
assigned_language = language_cols[index]

# Step 4: Prepare analysis dataframe
analysis_df = df[['Month', assigned_language]].copy()
analysis_df.rename(columns={assigned_language: 'Popularity'}, inplace=True)

# Step 5: Growth rate
analysis_df['Growth_Rate'] = analysis_df['Popularity'].pct_change() * 100

# Step 6: Smoothing
analysis_df['Moving_Avg'] = analysis_df['Popularity'].rolling(window=6).mean()
analysis_df['Moving_STD'] = analysis_df['Popularity'].rolling(window=6).std()

# Fill NaN
analysis_df.fillna(0, inplace=True)

# Step 7: Stats using NumPy
mean_growth = np.mean(analysis_df['Growth_Rate'])
std_growth = np.std(analysis_df['Growth_Rate'])

# Step 8: Lifecycle classification
conditions = [
    (analysis_df['Growth_Rate'] > 0) & (analysis_df['Growth_Rate'] < mean_growth),
    (analysis_df['Growth_Rate'] > mean_growth),
    (np.abs(analysis_df['Growth_Rate']) <= 1),
    (analysis_df['Growth_Rate'] < 0) & (analysis_df['Growth_Rate'] < (-1 * std_growth))
]

choices = ['Introduction', 'Growth', 'Maturity', 'Decline']

analysis_df['Lifecycle_Phase'] = np.select(conditions, choices, default='Maturity')

# Step 9: Summary counts
phase_counts = analysis_df['Lifecycle_Phase'].value_counts()
total = len(analysis_df)

phase_percent = (phase_counts / total) * 100

# Step 10: Dominant phase
dominant_phase = phase_counts.idxmax()

# Step 11: Outputs
print(f"Student Name: {STUDENT_NAME}")
print(f"Assigned Language: {assigned_language}")
print("=" * 50)

print("\nLifecycle Classification Table")
print(analysis_df[['Month', 'Popularity', 'Growth_Rate', 'Moving_Avg', 'Moving_STD', 'Lifecycle_Phase']].head(12))

print("\nPhase Counts")
print(phase_counts)

print("\nPhase Percentages")
print(phase_percent.round(2))

print("\nDominant Stage")
print(f"{assigned_language} is currently in the {dominant_phase} phase.")

# Step 12: Simple transition insight
transitions = analysis_df['Lifecycle_Phase'].shift() != analysis_df['Lifecycle_Phase']
transition_points = analysis_df[transitions]

print("\nLifecycle Transitions (first few)")
print(transition_points[['Month', 'Lifecycle_Phase']].head(10))
```
