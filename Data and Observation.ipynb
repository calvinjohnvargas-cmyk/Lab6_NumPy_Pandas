```python
import numpy as np

# Step 1: Set up student credentials
STUDENT_NAME = "Calvin"
STUDENT_ID = 250698

print(f"--- {STUDENT_NAME}'s Array Filtering Analysis ---\n")

# Step 2: Set the random seed
np.random.seed(STUDENT_ID)

# Step 3: Generate the array
# The upper limit uses (250698 % 100) + 50 = 98 + 50 = 148
max_value = (STUDENT_ID % 100) + 50
x = np.random.randint(1, max_value, size=10)

# Step 4: Filter the array
# Lower bound: 250698 % 50 = 48
# Upper bound: 250698 % 100 = 98
# Filter condition: Keep numbers strictly greater than 48 AND strictly less than 98
lower_bound = STUDENT_ID % 50
upper_bound = STUDENT_ID % 100
filtered = x[(x > lower_bound) & (x < upper_bound)]

# Step 5: Print results
print(f"Original Array: {x}")
print(f"Filtered subset (>{lower_bound} and <{upper_bound}): {filtered}")
```

```python
import numpy as np

# Step 1: Set up student credentials
STUDENT_NAME = "Calvin"
STUDENT_ID = 250698

print(f"--- {STUDENT_NAME}'s Linear Algebra Analysis ---\n")

# Step 2: Set the random seed
np.random.seed(STUDENT_ID)

# Step 3: Generate two 2x2 matrices
# The upper limit uses (250698 % 50) + 20 = 48 + 20 = 68
max_value = (STUDENT_ID % 50) + 20
A = np.random.randint(1, max_value, size=(2, 2))
B = np.random.randint(1, max_value, size=(2, 2)) 

print("Matrix A:\n", A, "\n")
print("Matrix B:\n", B, "\n")

# Step 4: Perform matrix operations
product = A @ B  # Matrix multiplication (dot product)
det_A = np.linalg.det(A) # Determinant of Matrix A

# Step 5: Print results
print("A @ B (Matrix Multiplication):\n", product, "\n")
# Rounded to 2 decimal places because determinants often become floats
print(f"Determinant of A: {det_A:.2f}")
```
